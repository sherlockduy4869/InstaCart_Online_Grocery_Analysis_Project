```python
op_prior = pd.read_csv(DATA_SOURCE + "order_products__prior.csv")
```

```python
op_train = pd.read_csv(DATA_SOURCE + "order_products__train.csv")
```

```python
orders = pd.read_csv(DATA_SOURCE + "orders.csv")
```

```python
orders_eval_prior = orders[orders["eval_set"] == "prior"].copy()
```

```python
orders_eval_prior = orders_eval_prior.sort_values(by = "user_id")
```

```python
orders_eval_train = orders[orders["eval_set"] == "train"].copy()
```

```python
orders_eval_train = orders_eval_train.sort_values(by = "user_id")
```

```python
orders_eval_prior.head(10)
```

```python
#orders_eval_train.head(10)
```

```python
orders_eval_prior_by_user_id = orders_eval_prior.loc[orders_eval_prior["user_id"] == 7]
```

```python
orders_eval_train_by_user_id = orders_eval_train.loc[orders_eval_train["user_id"] == 7]
```

```python
orders_eval_train_by_user_id.head(10)
```

```python
df_extract_features = pd.merge(op_prior, orders_eval_prior_by_user_id, on = "order_id", how = "right")
```

```python
#FEATURE BY USER
```

```python
user_avg_reordered = (
    df_extract_features
    .groupby(["user_id"])["reordered"]
    .mean()
    .reset_index()
    .rename(columns = {"reordered":"user_avg_reordered"})
)
```

```python
user_avg_order_dow = (
    df_extract_features
    .groupby(["user_id"])["order_dow"]
    .mean()
    .reset_index(name="user_avg_order_dow")
)
```

```python
user_avg_order_hour_of_day = (
    df_extract_features
    .groupby(["user_id"])["order_hour_of_day"]
    .mean()
    .reset_index(name="user_avg_order_hour_of_day")
)
```

```python
user_avg_day_since_prior_order = (
    df_extract_features
    .groupby(["user_id"])["days_since_prior_order"]
    .mean()
    .reset_index(name="user_avg_day_since_prior_order")
)
```

```python
user_total_orders = (
    df_extract_features
    .groupby(["user_id"])["order_number"]
    .max()
    .reset_index(name="user_total_orders")
)
```

```python
#FEATURE BY USER & PRODUCT
```

```python
up_avg_reordered = (
    df_extract_features
    .groupby(["user_id", "product_id"])["reordered"]
    .mean()
    .reset_index(name="up_avg_reordered")
)
```

```python
up_total_order = (
    df_extract_features
    .groupby(["user_id", "product_id"])["product_id"]
    .size()
    .reset_index(name="up_total_order")
)
```

```python
up_avg_order_dow = (
    df_extract_features
    .groupby(["user_id", "product_id"])["order_dow"]
    .size()
    .reset_index(name="up_avg_order_dow")
)
```

```python
up_avg_order_hour_of_day = (
    df_extract_features
    .groupby(["user_id", "product_id"])["order_hour_of_day"]
    .size()
    .reset_index(name="up_avg_order_hour_of_day")
)
```

```python
up_avg_cart_position = (
    df_extract_features
    .groupby(["user_id", "product_id"])["add_to_cart_order"]
    .size()
    .reset_index(name="up_avg_cart_position")
)
```

```python
up_avg_day_since_prior_order = (
    df_extract_features
    .groupby(["user_id", "product_id"])["days_since_prior_order"]
    .size()
    .reset_index(name="up_avg_day_since_prior_order")
)
```

```python
#FEATURE BY PRODUCTS
```

```python
product_total_orders = (
    df_extract_features
    .groupby(["product_id"])["order_id"]
    .size()
    .reset_index(name="product_total_orders")
)
```

```python
product_avg_reordered = (
    df_extract_features
    .groupby(["product_id"])["reordered"]
    .mean()
    .reset_index(name="product_avg_reordered")
)
```

```python
product_avg_cart_position = (
    df_extract_features
    .groupby(["product_id"])["add_to_cart_order"]
    .mean()
    .reset_index(name="product_avg_cart_position")
)
```

```python
df_train = df_extract_features
```

```python
df_train = df_train.merge(up_avg_reordered, on = ["user_id", "product_id"], how = "left")
df_train = df_train.merge(up_total_order, on = ["user_id", "product_id"], how = "left")
df_train = df_train.merge(up_avg_order_dow, on = ["user_id", "product_id"], how = "left")
df_train = df_train.merge(up_avg_order_hour_of_day, on = ["user_id", "product_id"], how = "left")
df_train = df_train.merge(up_avg_cart_position, on = ["user_id", "product_id"], how = "left")
df_train = df_train.merge(up_avg_day_since_prior_order, on = ["user_id", "product_id"], how = "left")
```

```python
df_train = df_train.merge(user_avg_reordered, on = "user_id", how = "left")
df_train = df_train.merge(user_avg_order_dow, on = "user_id", how = "left")
df_train = df_train.merge(user_avg_order_hour_of_day, on = "user_id", how = "left")
df_train = df_train.merge(user_avg_day_since_prior_order, on = "user_id", how = "left")
df_train = df_train.merge(user_total_orders, on = "user_id", how = "left")
```

```python
df_train = df_train.merge(product_total_orders, on = "product_id", how = "left")
df_train = df_train.merge(product_avg_reordered, on = "product_id", how = "left")
df_train = df_train.merge(product_avg_cart_position, on = "product_id", how = "left")
```

```python
df_train.isna().sum()
```

```python
df_train["days_since_prior_order"] = df_train["days_since_prior_order"].fillna(0)
```

```python
df_train.isna().sum()
```

```python
df_train = df_train.sort_values(by = ["user_id", "order_number"])
```

```python
df_test = pd.merge(op_train, orders_eval_train_by_user_id, on = "order_id", how = "right")
```

```python
df_test = df_test[df_test["reordered"] == 1]
```

```python
df_test = df_test.merge(up_avg_reordered, on = ["user_id", "product_id"], how = "left")
df_test = df_test.merge(up_total_order, on = ["user_id", "product_id"], how = "left")
df_test = df_test.merge(up_avg_order_dow, on = ["user_id", "product_id"], how = "left")
df_test = df_test.merge(up_avg_order_hour_of_day, on = ["user_id", "product_id"], how = "left")
df_test = df_test.merge(up_avg_cart_position, on = ["user_id", "product_id"], how = "left")
df_test = df_test.merge(up_avg_day_since_prior_order, on = ["user_id", "product_id"], how = "left")
```

```python
df_test = df_test.merge(user_avg_reordered, on = "user_id", how = "left")
df_test = df_test.merge(user_avg_order_dow, on = "user_id", how = "left")
df_test = df_test.merge(user_avg_order_hour_of_day, on = "user_id", how = "left")
df_test = df_test.merge(user_avg_day_since_prior_order, on = "user_id", how = "left")
df_test = df_test.merge(user_total_orders, on = "user_id", how = "left")
```

```python
df_test = df_test.merge(product_total_orders, on = "product_id", how = "left")
df_test = df_test.merge(product_avg_reordered, on = "product_id", how = "left")
df_test = df_test.merge(product_avg_cart_position, on = "product_id", how = "left")
```

```python
df_test["up_avg_reordered"] = df_test["up_avg_reordered"].fillna(0)
df_test["up_total_order"] = df_test["up_total_order"].fillna(0)
df_test["up_avg_order_dow"] = df_test["up_avg_order_dow"].fillna(0)
df_test["up_avg_order_hour_of_day"] = df_test["up_avg_order_hour_of_day"].fillna(0)
df_test["up_avg_cart_position"] = df_test["up_avg_cart_position"].fillna(0)
df_test["up_avg_day_since_prior_order"] = df_test["up_avg_day_since_prior_order"].fillna(0)
df_test["product_total_orders"] = df_test["product_total_orders"].fillna(0)
df_test["product_avg_reordered"] = df_test["product_avg_reordered"].fillna(0)
df_test["product_avg_cart_position"] = df_test["product_avg_cart_position"].fillna(0)
```
