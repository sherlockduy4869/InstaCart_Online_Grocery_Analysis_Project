```python
import pandas as pd               # For data manipulation
import numpy as np                # For numerical operations
import matplotlib.pyplot as plt   # For basic plotting
import seaborn as sns             # For pretty plotting

# This makes sure we'll have the graphs in the notebook
%matplotlib inline
```

```python
# Loading the csv files into variables (DataFrames)
aisles = pd.read_csv('aisles.csv')
departments = pd.read_csv('departments.csv')
orders = pd.read_csv('orders.csv')
products = pd.read_csv('products.csv')
op_prior = pd.read_csv('order_products__prior.csv') 
op_train = pd.read_csv('order_products__train.csv')

print("All data loaded successfully!")
```

```python
# I choose to check the Orders table first
print("ORDERS TABLE:")
display(orders.head())
```

```python
# Second check the Products table
print("PRODUCTS TABLE:")
display(products.head())
```

```python
# Third check the orders Prior since this is the original dataset(I feel like this did not work, but okay)
print("ORDERS PRIOR TABLE:")
display(products.head())
```

```python
# Third check the orders Prior since this is the original dataset(I feel like this is redundant)
print("ORDERS TRAIN TABLE:")
display(products.head())
```

```python
# I prefer to start the merge of items (prior) with product details =This is like a SQL LEFT JOIN or VLOOKUP on excel
merged_prior = pd.merge(op_prior, products, on='product_id', how='left')
```

```python
# 2. I include a merge with the departments and aisles (optional, but I think is important to see how they are placed)
merged_prior = pd.merge(merged_prior, aisles, on='aisle_id', how='left')
merged_prior = pd.merge(merged_prior, departments, on='department_id', how='left')
```

```python
# 3. Last merge is with the Orders table to get the User_ID and Time info
# I choose to merge on 'order_id'
full_data = pd.merge(merged_prior, orders, on='order_id', how='left')
```

```python
# Just to check if it worked, I look at the first few rows of our Master Table
display(full_data.head())
```

```python
# First thought, check when people buy and what they buy
#1. Orders by day of Week (0-6)
plt.figure(figsize=(10,5))
sns.countplot(x='order_dow', data=orders, color='teal')
plt.title('Busiest Days of the Week')
plt.xlabel('Day of Week (0 = Sunday/Saturday depending on dataset)')
plt.ylabel('Total Orders')
plt.show()
```

```python
#Day 0 is typically Sunday and Day 1 is Monday. I think this means customers are doing their "weekly restock" immediately as the week starts. 
#Day 0 will be our highest probability for large orders
```

```python
# 2. Orders by Hour of Day
plt.figure(figsize=(10,5))
sns.countplot(x='order_hour_of_day', data=orders, color='orange')
plt.title('Busiest Hours of the Day')
plt.xlabel('Hour (24h format)')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Busiest hours are 10:00 (10 AM) and 16:00 (4 PM).
#Customers prefer to shop at daytime with little activity in the night.
```

```python
# Step 2. What are our best sellers?
# This should count the occurrences of product names in our Master Table, without an overlap in names
top_products = full_data['product_name'].value_counts().head(10)

plt.figure(figsize=(10,6))
# I wanted a horizontal table so I swap x and y here so I can have a Horizontal Bar Chart
sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')

plt.title('Top 10 Most Ordered Products')
plt.xlabel('Number of Orders')
plt.ylabel('Product Name')
plt.show()
```

```python
#Our top 10 products are all fresh produce (fruits/vegetables)
#Primary products are fresh produce
```

```python
# Next check is the reorder levels
# 1. If I group by product and calculate two things:
#    - count: How many times was it bought total?
#    - mean of 'reordered': What % of those times was it a reorder?
product_stats = full_data.groupby('product_name').agg({
    'reordered': ['count', 'mean']
})
```

```python
# 2. Rename the columns to avoid confusion
product_stats.columns = ['total_orders', 'reorder_probability']
```

```python
# 3. Next is sort by 'reorder_probability' to see the "Most reordered" items first
top_reorders = product_stats.sort_values('reorder_probability', ascending=False)

# 4. create a display copy
display_table = top_reorders.head(15).copy()
display_table['reorder_probability'] = (display_table['reorder_probability'] * 100).round(2).astype(str) + '%'

print("Top 15 Products with the Highest Reorder Rate:")
display(display_table)
```

```python
#Raw Veggie wrappers is a favorite to the customers
#Orange Energy Shorts could be statistical noise(If it is only 1 customer making the order 13 times. We need to recheck
#Bananas have 400,000+ orders, but lower reorder rate= It brings traffic / high business volume but Serenity Pads' and 'Veggie Wrappers have high customer loyalty/reorder rates
```

```python
# OPTIONAL: The specific stats for Bananas specifically since it is the best seller
print("\nStats for Bananas:")
banana_stats = product_stats[product_stats.index.str.contains("Banana")]
display(banana_stats)
```

```python
# I want to add a filter for products with at least 10,000 orders to remove "niche" noise(veggie wrappers, e.t.c)
power_products = product_stats[product_stats['total_orders'] > 10000].copy()

# I will sort by reorder probability
top_power_products = power_products.sort_values('reorder_probability', ascending=False).head(10)

print("Top 10 High-Volume Essentials (The 'True' Leaders):")
display(top_power_products)
```

```python
# Specific check Bananas vs. Wrappers
print("\nDirect Comparison:")
target_items = ['Banana', 'Raw Veggie Wrappers', 'Orange Energy Shots']
display(product_stats.loc[product_stats.index.isin(target_items)])
```

```python
#Next step Segmented Analysis by DS
# Grouping by Department and calculating reorder rate
dept_stats = full_data.groupby('department').agg({
    'reordered': 'mean',
    'order_id': 'count'
}).reset_index()

dept_stats.columns = ['department', 'reorder_rate', 'total_orders']
```

```python
# Filter for departments with significant volume (just to see if it makes a difference)
dept_stats = dept_stats[dept_stats['total_orders'] > 1000]
```

```python
# Sorting and Plotting
dept_stats = dept_stats.sort_values('reorder_rate', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x='reorder_rate', y='department', data=dept_stats, palette='coolwarm')
plt.title('Which Departments have the Most Loyal Customers?')
plt.xlabel('Reorder Rate (Probability of buying again)')
plt.axvline(full_data['reordered'].mean(), color='k', linestyle='--', label='Global Average')
plt.legend()
plt.show()
```

```python
#The most reordered items are those that expire or quickly bought by many customers (Milk, Eggs, Bananas).
#The lowest items (Shampoo, Spices) are bought in bulk and last for months. So these make stock-up products and explains the low reorder rate
#We must create a feature called department_reorder_rate so that our model learns that not buying milk is a warning sign, but not buying shampoo is normal.
```

```python
#How frequently do they reorder? Is there any relationship between "Days Since Last Order" and "Reorder Probability"
# Group by the number of days since the last order
frequency_stats = full_data.groupby('days_since_prior_order').agg({
    'reordered': 'mean'
}).reset_index()

# 2. Plot the trend
plt.figure(figsize=(12,5))
sns.lineplot(x='days_since_prior_order', y='reordered', data=frequency_stats, marker='o', color='crimson')
plt.title('Does Time Affect Loyalty? (Reorder Rate vs. Days Since Last Shop)')
plt.xlabel('Days Since Prior Order')
plt.ylabel('Probability of Reordering')
plt.grid(True)
plt.show()
```

```python
#My understanding of the trend
#Day 0 Spike (0.74) = Means this is probably something the customer forgot to buy and they make the order immediately
#Day 7 Peak and the small ones at #Day 14 and #Day 21
#After Day 7, the line trends downward. If a customer waits 30 days, their probability of reordering staples/usuals drops to ~45%

#The customers have a week/7 day cycles. They will reorder in 7 days. 
#If they wait longer like to 30 days, they may be doing a restocking instead of reorder.
#These customers couuld have made orders elsewhere.
```

```python
# Customer Habits: Do all the customers make similar reorders, or just new customers all the time?
#Calculate the reorder ratio for EACH user_id
user_loyalty = full_data.groupby('user_id')['reordered'].mean().reset_index()
user_loyalty.columns = ['user_id', 'user_reorder_ratio']

# 2. Plot the distribution of user_IDs
plt.figure(figsize=(10,5))
sns.histplot(user_loyalty['user_reorder_ratio'], bins=30, kde=True, color='green')
plt.title('User Types: Explorers (Left) vs. Creatures of Habit (Right)')
plt.xlabel('Average Reorder Rate per User')
plt.ylabel('Number of Users')
plt.show()
```

```python
# The customer segment almost equally at 0.4-0.5 distribution. 50% staples (Milk, Bread) and 50% new desires (Dinner specific ingredients)
# The ones at the 0.8-1.0 represent those that rarely change their diets
#We should probably categorize users_ids into segments. This way our model will likely find it very easy to predict the non-changing users and we can focus on making our model better to predict the "Explorers.
#Creatures of habit= those that have predictable behaviour and the model will pick these
```

```python
# Group by User_ID and Eval_Set, then count how many orders exist in each
split_check = orders.groupby(['user_id', 'eval_set'])['order_id'].count().reset_index()

# Rename for ease of use
split_check.columns = ['user_id', 'dataset_type', 'order_count']

# Check to see the distribution
print("How many orders per user in each dataset?")
print(split_check.groupby('dataset_type')['order_count'].describe())
```

```python
# Visualize it
plt.figure(figsize=(8,5))
sns.boxplot(x='dataset_type', y='order_count', data=split_check)
plt.title('Distribution of Order Counts: Prior (History) vs. Train (Target)')
plt.show()
```

```python
# Are they consistent in the reorder?
# Calculation of reorder rate for Prior (History)
prior_reorder_rate = op_prior['reordered'].mean()

# Calculation of reorder rate for Train (The Target)
train_reorder_rate = op_train['reordered'].mean()

print(f"Reorder Rate in History (Prior): {prior_reorder_rate:.2%}")
print(f"Reorder Rate in Target (Train):  {train_reorder_rate:.2%}")
```

```python
# Simple Bar Chart Comparison
plt.figure(figsize=(6,5))
sns.barplot(x=['Prior (History)', 'Train (Target)'], y=[prior_reorder_rate, train_reorder_rate], palette='Blues')
plt.title('Are Reorder Rates Consistent?')
plt.ylim(0, 1) # Set limit to 0-100%
plt.ylabel('Reorder Probability')
plt.show()
```

```python
#The reorder rates are almost identical (~59% vs ~60%), meaning the prior is a valid predictor of the future orders
```

```python
#WORK IN PROGRESS
# Check what is bought together
#Merge the order items with themselves on 'order_id'
df_pairs = op_prior.merge(op_prior, on='order_id')

#Filter out "Self-Joins" (Banana + Banana)
df_pairs = df_pairs[df_pairs['product_id_x'] < df_pairs['product_id_y']]

# 3. Get the names as a small dictionary for the products
product_name_map = products.set_index('product_id')['product_name'].to_dict()

df_pairs['product_x'] = df_pairs['product_id_x'].map(product_name_map)
df_pairs['product_y'] = df_pairs['product_id_y'].map(product_name_map)

# 4. Count the top pair combinations in the orders
top_pairs = df_pairs.groupby(['product_x', 'product_y']).size().reset_index(name='frequency')
top_pairs = top_pairs.sort_values('frequency', ascending=False).head(10)

print("Top 10 Most Common Product Combinations:")
display(top_pairs)
```
