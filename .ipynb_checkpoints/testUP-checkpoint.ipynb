```python
import pandas as pd
from collections import defaultdict

# ======================
# CONFIG
# ======================
MAX_ORDERS = 20000
OUTPUT_FILE = "instacart_upgrowth.txt"

# ======================
# LOAD DATA
# ======================

DATA_SOURCE = "/Users/mac/Desktop/DSTI_Project/Machine_Learning/Project 2/archive/"

orders = pd.read_csv(DATA_SOURCE + "orders.csv")
order_products = pd.read_csv(DATA_SOURCE + "order_products__prior.csv")

orders = orders.head(MAX_ORDERS)
order_products = order_products[
    order_products["order_id"].isin(orders["order_id"])
]

# ======================
# BUILD TRANSACTIONS
# ======================
transactions = defaultdict(list)

for _, row in order_products.iterrows():
    # quantity = 1, profit = 1 (giả lập)
    utility = 1
    transactions[row["order_id"]].append(
        (int(row["product_id"]), utility)
    )

# ======================
# WRITE SPMF FILE
# ======================
with open(OUTPUT_FILE, "w") as f:
    for items in transactions.values():
        item_ids = [str(i) for i, _ in items]
        utils = [str(u) for _, u in items]
        TU = sum(int(u) for u in utils)

        line = (
            " ".join(item_ids)
            + " : "
            + str(TU)
            + " : "
            + " ".join(utils)
        )
        f.write(line + "\n")

print("✔ File saved:", OUTPUT_FILE)
```
